Questão 5

Qual é a diferença, em termos de endereço, conteúdo e total de bits, entre as seguintes organizações de MP:

a) memória A: 32K células de 8 bits cada;

b) memória B: 16K células de 16 bits cada; 

c) memória C: 16K células de 8 bits cada?

Dados:

•	a) Memória A: 32K células × 8 bits

•	b) Memória B: 16K células × 16 bits

•	c) Memória C: 16K células × 8 bits

Análise:

•	Endereço: Depende da quantidade de células.

o	32K = 2¹⁵ → endereço de 15 bits

o	16K = 2¹⁴ → endereço de 14 bits

•	Conteúdo: Tamanho de cada célula

o	A e C: 8 bits (1 byte)

o	B: 16 bits (2 bytes)

•	Total de bits: Quantidade × tamanho da célula

o	A: 32768 × 8 = 262.144 bits

o	B: 16384 × 16 = 262.144 bits

o	C: 16384 × 8 = 131.072 bits

Diferenças:

•	A tem endereço maior (15 bits) que B e C (14 bits).

•	B armazena 16 bits por célula, A e C armazenam 8 bits.

•	A e B têm capacidade total igual, C tem metade da capacidade de A e B.


Questão 6

Qual é a função do registrador de endereços de memória (REM)? E do registrador de dados de memória (RDM)?

•	REM (Registrador de Endereços de Memória): Armazena o endereço da célula de memória que será acessada. Envia esse endereço ao barramento de endereços para selecionar a posição correta.

•	RDM (Registrador de Dados de Memória): Armazena os dados que serão escritos na memória ou os dados lidos da memória. Funciona como buffer entre o processador e a memória, temporariamente retendo as informações durante a transferência.



Questão 7

Descreva os barramentos que interligam processador e MP, indicando função e direção do fluxo de sinais de cada um.

Barramentos entre processador e memória principal:

1.	Barramento de Endereços

o	Função: Transporta o endereço da célula a ser acessada.

o	Direção: Unidirecional → do processador para a memória.

2.	Barramento de Dados

o	Função: Transporta os dados que serão lidos ou escritos.

o	Direção: Bidirecional → do processador para a memória (escrita) e da memória para o processador (leitura).

3.	Barramento de Controle

o	Função: Transporta sinais de comando (leitura, escrita, habilitação).

o	Direção: Principalmente unidirecional → do processador para a memória; alguns sinais de status podem retornar.

Questão 33

Você considera válida a afirmação “um computador com mais poder de processamento pode armazenar mais programas”?

Afirmação: “Um computador com mais poder de processamento pode armazenar mais programas”

Resposta: Inválida.

O poder de processamento (velocidade, número de núcleos) não está ligado à capacidade de armazenamento. A quantidade de programas depende do tamanho da memória e do disco, não da velocidade do processador.


Questão 34

Você considera válida a afirmação “vale aumentar a capacidade da memória principal para que o acesso aos meios magnéticos (discos rígidos e disquetes) seja mais rápido”?

Afirmação: “Aumentar a memória principal torna acesso a discos e disquetes mais rápido”

Resposta: Parcialmente válida.

Mais memória permite armazenar mais dados/programas na RAM, reduzindo a necessidade de ler do disco. Mas não aumenta a velocidade do disco em si — apenas diminui a frequência de acesso.


Questão 8

Descreva passo a passo uma operação de leitura. Utilize um diagrama esquemático.

Operação de leitura:

1.	Processador envia o endereço para o REM.

2.	Endereço vai ao barramento de endereços → memória decodifica.

3.	Processador envia sinal de Leitura pelo barramento de controle.

4.	Memória coloca os dados da célula no barramento de dados.

5.	Dados são transferidos para o RDM.

6.	Processador lê os dados do RDM.

(Diagrama: Processador → REM → Barramento Endereços → Memória → Barramento Dados → RDM → Processador)

Operação de escrita:

3.	Dados a gravar são colocados no RDM.

4.	Processador envia sinal de Escrita pelo barramento de controle.

5.	Dados do RDM vão ao barramento de dados.

6.	Memória grava os dados na célula indicada.

(Diagrama: Processador → REM → Barramento Endereços → Memória; Processador → RDM → Barramento Dados → Memória)


Questão 13

Considere uma memória com capacidade de armazenamento de 64K bytes; cada célula pode armazenar 1 byte de informação e cada caractere é codificado com 8 bits. Resolveu-se armazenar na memória deste sistema um conjunto de caracteres do seguinte modo. A partir do endereço (hexadecimal) 27FA, foram escritos sucessivamente grupos de 128 caracteres iguais, iniciando pelo grupo de As, seguido do grupo de Bs, e assim por diante. Qual deverá ser o endereço correspondente ao local onde está armazenado o “J”?

Dados:

•	64K bytes = 65536 células × 1 byte

•	Endereço inicial: 27FA₁₆

•	Grupos de 128 caracteres: A, B, ...

•	Procurar: 1.º caractere do grupo J

Cálculo:

128 = 2⁷ → cada grupo ocupa 128 endereços.

Letra A = 1.º grupo → J = 10.º grupo.

Deslocamento = 9 × 128 = 1152₁₀ = 480₁₆

Endereço = 27FA₁₆ + 480₁₆ = 2C7A₁₆


Questão 14

O custo das memórias SRAM é maior que o das memórias DRAM. No entanto, o processo de conexão das memórias DRAM é mais complexo que o das SRAM e, em consequência, o preço da interface das DRAM é bem maior que o das SRAM. Supondo que uma interface de DRAM custe R$5,00, uma interface de SRAM custe R$1,00, o preço por bit de uma SRAM é de R$0,00002 e o de uma DRAM é de R$0,00001, calcule quantos bits deve ter uma memória dinâmica (DRAM) para que o conjunto seja mais barato.

Dados:

•	Interface DRAM: R$5,00; por bit: R$0,00001

•	Interface SRAM: R$1,00; por bit: R$0,00002

•	Encontrar n (bits da DRAM) para custo igual.

Equação:

Custo DRAM = 5 + 0,00001 × n

Custo SRAM = 1 + 0,00002 × n

Igualar:

5 + 0,00001n = 1 + 0,00002n

4 = 0,00001n

n = 400.000 bits

Acima de 400.000 bits, a DRAM fica mais barata.
